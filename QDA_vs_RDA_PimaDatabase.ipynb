```python
#On crée une function permettant de renommer les modalités de la variable Outcome

def replace_Colstring(data, column, value_to_replace, replacement_value):
    data.loc[:, column] = data[column].map(lambda x: str(x).replace(value_to_replace,
                                                                    replacement_value))
    return data
```

```python
data=replace_Colstring(data, "Outcome","1", "Diabetes")
data=replace_Colstring(data, "Outcome","0", "Pas_Diabetes") 
    
```

```python
# L'insuline devrait être supérieur à zéro et toutes les autres var d'ailleurs
data = data[~(data == 0).any(axis=1)]
```

```python
# On réalise finalement l'ADQ
data_qda=data[["Outcome","BMI","Glucose","BloodPressure","SkinThickness"]]
data_qda.Outcome.value_counts()
```

```python
# QDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
#Standardisation des données
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import *

# Séparer les données en variables prédictives (X) et variables cibles (y)
X = data_qda.iloc[:,data_qda.columns != 'Outcome']
y = data_qda["Outcome"]
scaler = StandardScaler()
X = scaler.fit_transform(X)
# Séparer les données en ensembles d'apprentissage et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2023)

# Créer un objet ADQ
qda = QuadraticDiscriminantAnalysis()

# Adapter l'ADQ aux données d'apprentissage
qda.fit(X_train, y_train)

# Utiliser l'ADQ pour prédire les classes des données de test
y_pred = qda.predict(X_test)

# Calculer la précision de la prédiction
accuracy = accuracy_score(y_test, y_pred)

# Afficher la précision de la prédiction
print('Précision de la prédiction:', accuracy)
# Afficher le rapport de classification

print(classification_report(y_test, y_pred))
```

```python
# Vaildation croisée pour la RDA
from sklearn.model_selection import KFold, cross_val_score
# Définir la valeur du paramètre de régularisation
reg_params = np.arange(0.1, 1.05, 0.05)

# Définition de l'objet KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Dictionnaire pour stocker les performances pour chaque valeur de reg_params
perf_dict = {}

# Boucle sur les valeurs de reg_params
for reg_param in reg_params:
    # Création de l'objet RDA avec la valeur de reg_params
    rda = QuadraticDiscriminantAnalysis(reg_param=reg_param)
    
    # Calcul de la performance moyenne sur l'ensemble des folds avec cross_val_score
    perf = np.mean(cross_val_score(rda, X_train, y_train, cv=kf, scoring='accuracy'))
    
    # Stockage de la performance pour cette valeur de reg_param
    perf_dict[reg_param] = perf

# Identification de la valeur de reg_param qui donne la meilleure performance moyenne
best_reg_param = max(perf_dict, key=perf_dict.get)

# Création de l'objet RDA final avec la meilleure valeur de reg_param
rda = QuadraticDiscriminantAnalysis(reg_param=best_reg_param)

# Entraînement de l'objet RDA final sur l'ensemble des données d'entraînement
rda.fit(X_train, y_train)

# Évaluation des performances de l'objet RDA final sur l'ensemble des données de test
y_pred = rda.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# QDA learning curve
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    estimator=qda, X=X, y=y, cv=10,
    scoring='balanced_accuracy', train_sizes=np.linspace(0.1, 1.0, 10),
    shuffle=True, random_state=42)

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

plt.plot(train_sizes, train_scores_mean, label='Training score')
plt.plot(train_sizes, test_scores_mean, label='Cross-validation score')
plt.title('QDA Learning curve')
plt.xlabel('Taille des données')
plt.ylabel('Score')
plt.legend(loc='best')
plt.show()
```

```python
# RDA learning curve
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    estimator=rda, X=X, y=y, cv=10,
    scoring='balanced_accuracy', train_sizes=np.linspace(0.1, 1.0, 10),
    shuffle=True, random_state=42)

train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

plt.plot(train_sizes, train_scores_mean, label='Training score')
plt.plot(train_sizes, test_scores_mean, label='Cross-validation score')
plt.title('RDA Learning curve')
plt.xlabel('Taille des données')
plt.ylabel('Score')
plt.legend(loc='best')
plt.show()
```
